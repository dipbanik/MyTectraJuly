## Problem : Compare the populations of the 10 most populous countries in 2007

Us ecountries.csv

create bar graphs using plt.bar()
sort a data frame with sort_values()


```
import pandas as pd
from matplotlib import pyplot as plt

data = pd.read_csv('https://raw.githubusercontent.com/dipbanik/MyTectraJuly/master/visualization/countries.csv')
data.head()
```

```
data_2007 = data[data.year == 2007]
data_2007.head()
```

```
top10 = data_2007.sort_values('population', ascending = False).head(10)
#getting the top 10 most populous countries in the world. 
```

```
top10.head(10)
```

```
x = range(10)
# x = 0,1,2,3... 9
x
```

```
plt.bar(x, top10.population / 10**6)
plt.xticks(x, top10.country, rotation = 'vertical')
plt.ylabel('Population in millions')
plt.title('10 most populous countries')
plt.show()
```

## Compare the GDPs of the 10 most populous countries in 2007


```
plt.subplot(2,1,1)
plt.bar(x,top10.population / 10**6)
plt.xticks([], [])
plt.title('10 Most Populous Countries')
plt.legend(['Population in Millions'])

plt.subplot(2,1,2)
plt.bar(x, top10.gdpPerCapita * top10.population / 10 ** 9)
plt.xticks(x, top10.country, rotation='vertical')
plt.legend(['GDP in Billions'])
plt.show()
```
